```python
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread("/content/cat_image.jpg")
cv2_imshow(image)
```

```python
print("SHAPE:", image.shape)
# 2571 => height
# 3840 => width
# 3 => color channels (BGR) => Blue, Green, Red
print("HEIGHT:",image.shape[0])
print("WIDTH:",image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b, g, r) = image[100, 50] #rasmdagi 100-qator va 50-ustundagi ranglarni ko'rsatyapti
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray,(300,300))
cv2_imshow(resized)
```

```python
cropped = image[100:1700, 350:1800]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("cropped_cat.jpg", cropped)
```
